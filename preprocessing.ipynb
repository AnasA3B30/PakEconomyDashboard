```python
sugar_mill_stocks = pd.read_csv(r'/home/anas/Desktop/Economy Report/data/adam sugar mills stock 2002-2022.csv') # linux paths lol
sugar_mill_stocks #exploring sugar mill data because all company stocks files have the same format so this covers them
```

```python
sugar_mill_stocks.info()
```

```python
sugar_mill_stocks.describe()
```

```python
sugar_mill_stocks['Date']=pd.to_datetime(sugar_mill_stocks['Date']) # to create year, month, day columns
```

```python
sugar_mill_stocks['Year']=sugar_mill_stocks['Date'].dt.year
sugar_mill_stocks['Month']=sugar_mill_stocks['Date'].dt.month
sugar_mill_stocks['Day']=sugar_mill_stocks['Date'].dt.day
```

```python
sugar_mill_stocks
```

```python
years = np.arange(2001,2023) # value-month plots for years
for year in years:
    plot_data_raw = sugar_mill_stocks.loc[sugar_mill_stocks['Year']==year]
    plot_data_raw.sort_values('Day',inplace=True)
    plot_data = plot_data_raw.groupby('Month')[['Open','Low','High','Close',]].mean()
    volume = plot_data_raw.groupby('Month')['Volume'].sum()
    plt.figure(figsize=(18,15))
    plt.subplot(2,2,1)
    sns.lineplot(plot_data,marker='o')
    plt.title(year)
    plt.legend()
    plt.subplot(2,2,2)
    sns.barplot(plot_data)
    plt.title(f'Avg Yearly Values for {year}')
    plt.subplot(2,1,2)
    plt.title(f'Monthly volume traded for Year {year}')
    sns.barplot(volume)
    plt.show()
```

```python
volume_grouped = sugar_mill_stocks.groupby('Year')['Volume'].sum() # find total volume traded by year
volume_grouped
```

```python
plt.figure(figsize=(20,7))
sns.barplot(volume_grouped) # plotting total value against years
```
